Resampling of DF

```python
folder_path = "/Users/francopallitto/Documents/DS_projects/nispera/data/processed/final_data"
#chose the name of the file
file_name = "final_data"
data_file = folder_path + "/" + file_name + ".csv"
df = pd.read_csv(data_file)
df = df.drop(columns=["Unnamed: 0"])
#choose the minutes of the resampling
time_resample = "60min"
```

```python
df['datetime'] = pd.to_datetime(df['datetime'].str.strip(), format='%Y-%m-%dT%H:%M:%S')
```

```python
grouped = df.groupby(["ID","problem_class"]).resample(time_resample, on = "datetime").mean()
```

```python
#chose the folder were you want to save the fi
grouped.to_csv("/Users/francopallitto/Documents/DS_projects/nispera/data/interim/final_data_"+time_resample+".csv")
```
